```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
training_path = "plot_data/googlenet/run-version_3-tag-training_loss_epoch.csv"
val_path = "plot_data/googlenet/run-version_3-tag-val_loss.csv"
```

```python
train_df = pd.read_csv(training_path)
val_df = pd.read_csv(val_path)

train_df = train_df[["Step", "Value"]]
val_df = val_df[["Step", "Value"]]

df = pd.merge(train_df, val_df, how="inner", on=["Step"], suffixes = ('_train', '_val'))

TITLE = "GoogleNet"
FILE_NAME = "googlenet"

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.set_title(f"{TITLE} Learning Curve")

ax.plot(range(len(df["Step"])), df["Value_train"], label="training_loss")
ax.plot(range(len(df["Step"])), df["Value_val"], label="validation_loss")

ax.set_xlabel("Epochs")
ax.set_ylabel("MSE Error")

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.legend()
fig.savefig(fname=f"plot_data/{FILE_NAME}.jpg", format="jpg")
plt.show()
```
